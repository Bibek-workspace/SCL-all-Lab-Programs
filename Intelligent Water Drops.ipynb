```python
import numpy as np

# -----------------------------
# Problem Setup (TSP)
# -----------------------------
np.random.seed(0)
n_nodes = 5

# Distance matrix
dist = np.random.randint(10, 50, (n_nodes, n_nodes))
np.fill_diagonal(dist, 0)

# Initial soil on edges
soil = np.ones((n_nodes, n_nodes)) * 100

# -----------------------------
# Parameters
# -----------------------------
n_drops = n_nodes
iterations = 50

a_v = 1
b_v = 0.01
c_v = 1

a_s = 1
b_s = 0.01
c_s = 1

rho = 0.1  # global soil update factor

# -----------------------------
# Helper Functions
# -----------------------------
def probability(i, j, visited):
    if j in visited or i == j:
        return 0
    return 1 / (soil[i][j] + 1e-6)

def choose_next(i, visited):
    probs = []
    nodes = []
    
    for j in range(n_nodes):
        if j not in visited and j != i:
            p = probability(i, j, visited)
            probs.append(p)
            nodes.append(j)
    
    probs = np.array(probs)
    probs /= probs.sum()
    
    return np.random.choice(nodes, p=probs)

# -----------------------------
# Main IWD Algorithm
# -----------------------------
best_path = None
best_cost = float('inf')

for it in range(iterations):
    paths = []
    
    for drop in range(n_drops):
        visited = [np.random.randint(0, n_nodes)]
        velocity = 1.0
        local_soil = 0
        
        while len(visited) < n_nodes:
            i = visited[-1]
            j = choose_next(i, visited)
            
            # Update velocity
            velocity += a_v / (b_v + c_v * soil[i][j]**2)
            
            # Time to travel
            time = dist[i][j] / velocity
            
            # Soil update
            delta_soil = a_s / (b_s + c_s * time**2)
            soil[i][j] = (1 - rho) * soil[i][j] - rho * delta_soil
            
            local_soil += delta_soil
            visited.append(j)
        
        # Complete cycle (return to start)
        cost = sum(dist[visited[i]][visited[i+1]] for i in range(n_nodes-1))
        cost += dist[visited[-1]][visited[0]]
        
        paths.append((visited, cost))
        
        if cost < best_cost:
            best_cost = cost
            best_path = visited
    
    # Global soil update on best path
    for i in range(len(best_path)-1):
        soil[best_path[i]][best_path[i+1]] *= (1 - rho)
    
    print(f"Iteration {it+1}: Best Cost = {best_cost}")

# -----------------------------
# Output
# -----------------------------
print("\nFinal Best Path:", best_path)
print("Final Best Cost:", best_cost)
```
