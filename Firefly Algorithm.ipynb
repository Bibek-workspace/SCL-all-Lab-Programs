```python
# Aim: Optimize function using Firefly Algorithm

import numpy as np

def f(x):
    return x**2

fireflies = np.random.uniform(-10, 10, 10)

for _ in range(50):
    for i in range(len(fireflies)):
        for j in range(len(fireflies)):
            if f(fireflies[j]) < f(fireflies[i]):
                fireflies[i] += 0.3 * (fireflies[j] - fireflies[i])

print("Best solution:", min(fireflies, key=f))
```
